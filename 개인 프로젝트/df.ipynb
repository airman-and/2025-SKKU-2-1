```python
import numpy as np
import matplotlib.pyplot as plt

# Define sigmoid function and its derivative
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmoid_derivative(z):
    s = sigmoid(z)
    return s * (1 - s)

# Create input range
z_vals = np.linspace(-10, 10, 200)
sig_vals = sigmoid(z_vals)
sig_deriv_vals = sigmoid_derivative(z_vals)

# Simulate gradient vectors in 2D for sigmoid and compare with ideal case
X, Y = np.meshgrid(np.linspace(-2, 2, 20), np.linspace(-2, 2, 20))
U = sigmoid_derivative(X) * 1   # dx direction
V = sigmoid_derivative(Y) * 1   # dy direction

# Create ideal gradient directions (uniform in all directions)
U_ideal = X
V_ideal = Y
norm = np.sqrt(U_ideal**2 + V_ideal**2)
U_ideal /= norm
V_ideal /= norm

# Plotting
plt.figure(figsize=(14, 6))

# Plot sigmoid and its derivative
plt.subplot(1, 2, 1)
plt.plot(z_vals, sig_vals, label="Sigmoid")
plt.plot(z_vals, sig_deriv_vals, label="Sigmoid Derivative")
plt.title("Sigmoid Function and its Derivative")
plt.xlabel("z")
plt.ylabel("Value")
plt.grid(True)
plt.legend()

# Plot vector field comparison
plt.subplot(1, 2, 2)
plt.quiver(X, Y, U, V, color='red', alpha=0.6, label="Sigmoid Gradient Directions")
plt.quiver(X, Y, U_ideal, V_ideal, color='blue', alpha=0.4, label="Optimal Gradient Directions")
plt.title("Gradient Direction: Sigmoid vs Optimal")
plt.xlabel("Weight w1")
plt.ylabel("Weight w2")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
